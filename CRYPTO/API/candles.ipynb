```python
import requests  # Install requests module first.
import pandas as pd
import datetime as dt
import time
```

```python
url = "https://public.coindcx.com/market_data/candles?pair=I-BTC_INR&interval=5m&limit=10"  # Replace 'B-BTC_USDT' with your desired market pair.
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
df['date and time'] = df['time'].apply(lambda x :time. strftime("%Y-%m-%d %H:%M:%M", time.localtime(int(str(x)[:-3]))))
```

```python
df = df.reset_index(drop=True)
```

```python
df = df[['date and time', 'open', 'high', 'low', 'close', 'volume']]

```
